```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fun(x):
    return 1 / (1 + x*x)
```

```python
total_x = []
for n in range(2, 11, 2):
    x = []
    for k in range(n + 1):
        x.append(-5 + 10 / n * k)
    total_x.append(x)
```

```python
X = []
n = 2
# for i in range(len(total_x)):
for j in range(len(total_x[0])):
    tmp_x = [1]
    for k in range(n):
        tmp_x.append(np.power(total_x[0][j], k + 1))
    X.append(tmp_x)
```

```python
Y = fun(np.array(total_x[0]))
```

```python
a = np.linalg.inv(np.array(X)) @ Y
```

```python
def fun_n2(x):
    ans = 0
    for i in range(len(a)):
        ans += a[i] * np.power(x, i)
    return ans
```

```python
x0 = np.arange(-5, 5, 0.1)  # Sample data.
x1 = np.array(total_x[0])
# 生成figure和axes实例
fig, ax = plt.subplots(figsize=(5, 5), layout='constrained')
# fig, axs = plt.subplots(2, 1, layout='constrained') # 画多张子图

# 画几条曲线
ax.plot(x0, fun(x0), label='quadratic')
ax.plot(x0, fun_n2(x0), label='qe')
# ax.scatter(x, fun(x)) # 散点图

# 设置图的属性
ax.set_xlabel('x label')  # x-label
ax.set_ylabel('y label')  # y-label
ax.set_title("Simple Plot")  # title
ax.legend(loc="lower left");  # 显示图例，对应ax.plot()中的lable属性, loc指定图例显示在图的哪个位置
ax.grid()

# 坐标区间控制
# ax.axis([-5, 5, 0, 1]) # 一次性设置xmin, xmax, ymin, ymax
# ax.set_xlim(-5, 5) # 单独设置x轴
# ax.set_ylim((0, 1)) # 单独设置y轴

#坐标轴刻度
# ax.set_xticks([-5, 0, 5]) # 设置x刻度为指定值
# ax.set_yticks([0,0.5,1], ['zero', '50', 'hundred']) # 在y轴刻度指定刻度显示指定文字

#添加注释
# ax.text(4.5, 40,'Sactter')
# ax.annotate('annotate cubic', xy=(4.2, 80), xytext=(5, 60),
#             arrowprops=dict(facecolor='black', shrink=0.05))
```
